#### Sorting

```python
# Sort the data by TotalSales in asc order
stores.sort_values( by = 'TotalSales', ascending = True ).head(2)
```

```python
# Sort the data by TotalSales in desc order
stores.sort_values( by = 'TotalSales', ascending = False, ignore_index = False ).head(2)
```

```python
# Sort the data by Location, TotalSales in asc order
# Sort the data by TotalSales, Location in asc order - Not the same as previous statement
stores.sort_values( by = ['Location', 'TotalSales'], ascending = True ).head(2)
```

```python
stores.sort_values( by = ['TotalSales', 'Location'], ascending = True ).head(2)
```

```python
# Sort the data by Location, TotalSales in desc order
stores.sort_values( by = ['Location', 'TotalSales'], ascending = False ).head(2)
```

```python
# Sort the data by Location in asc and TotalSales in desc order
stores.sort_values( by = ['Location', 'TotalSales'], ascending = [True, False], inplace = True )
```

```python
# sort the data back in the original / previous order
stores.sort_index( axis = 0 ).sort_index( axis = 1 ).head(2)
```
